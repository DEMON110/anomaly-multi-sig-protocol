```python
# Step 1: Install required libraries (if not already available)
!pip install scikit-learn pandas matplotlib

# Step 2: Upload your file in Colab
from google.colab import files
uploaded = files.upload()

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve

# Step 3: Load dataset (with tab separator handling)
file_name = list(uploaded.keys())[0]  # take uploaded file name
df = pd.read_csv(file_name, sep="\t")

# Step 4: Extract true labels and predicted scores
y_true = df['y_true'].astype(int)
y_score = df['y_score']

# Step 5: Calculate precision, recall, thresholds
precision, recall, thresholds = precision_recall_curve(y_true, y_score)

# Step 6: Calculate F1 scores
f1_scores = 2 * (precision * recall) / (precision + recall + 1e-10)

# Step 7: Find best threshold and F1
best_idx = np.argmax(f1_scores)
best_threshold = thresholds[best_idx] if best_idx < len(thresholds) else 0.5
best_f1 = f1_scores[best_idx]

print("Best Threshold:", best_threshold)
print("Best F1 Score:", best_f1)

# Step 8: Plot F1 vs Threshold
plt.figure(figsize=(8,5))
plt.plot(thresholds, f1_scores[:-1], label="F1 Score")
plt.axvline(x=best_threshold, color='r', linestyle='--', label=f"Best Threshold: {best_threshold:.4f}")
plt.xlabel("Threshold")
plt.ylabel("F1 Score")
plt.title("F1 Score vs Threshold")
plt.legend()
plt.show()
```
